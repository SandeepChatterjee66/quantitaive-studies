```python
import os
import pandas as pd
```

```python
path = os.getcwd()+'/dataset'
files = os.listdir(path)
#files
```

```python
os.chdir(path)
files_xls = [f for f in files if f[-3:] == 'xls']
files_xls
```

```python
dfs = []
```

```python
for f in files_xls:
    data = pd.read_excel(f)
    dfs.append(data)
```

```python
df = pd.concat(dfs)
```

```python
df.head()
```

```python
df.columns
```

```python
df['Addresses']
```

```python
df['Addresses'].tolist()[0:5]
```

```python
afulladdrs = df['Addresses'].dropna().tolist()
```

```python
name ='''[Gupta, Pranshu K.; Ranganath, Kalluri V. S.; Mishra, Lallan] Banaras Hindu Univ, Dept Chem, Inst Sci, Varanasi 221005, Uttar Pradesh, India; [Dubey, Nawal K.] Banaras Hindu Univ, Dept Bot, Inst Sci, Varanasi 221005, Uttar Pradesh, India'''
name
```

```python
import re
```

```python
re.sub("[\[].*?[\]]", "", name)
```

```python
addrs = []

for addr in afulladdrs:
    addr = re.sub("[\[].*?[\]]", "", addr)
    addrs.extend([name.strip() for name in str(addr).split(';')])
addrs
```

```python
def country_name(address):
    return address.split(',')[-1].strip().title().split(' ')[-1].strip()
```

```python
countries = [country_name(address) for address in addrs]
countries
```

```python
from collections import Counter
```

```python
countries_count = Counter(countries)
countries_count
```

```python
countrynames = countries_count.keys()
counts = [countries_count[c] for c in countrynames]

countrywise_df = pd.DataFrame({"Country" : countrynames, "Occurrences" : counts})
```

```python
countrywise_df
```

```python
country_df = countrywise_df.sort_values(by='Occurrences',ascending=False)
country_df.head(130)
```
