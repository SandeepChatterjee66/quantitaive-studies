```python
import os
import pandas as pd
```

```python
path = os.getcwd()+'/dataset'
files = os.listdir(path)
#files
```

```python
os.chdir(path)
files_xls = [f for f in files if f[-3:] == 'xls']
files_xls
```

```python
dfs = []
```

```python
for f in files_xls:
    data = pd.read_excel(f)
    dfs.append(data)
```

```python
df = pd.concat(dfs)
```

```python
df.head()
```

```python
df.columns
```

```python
df['Author Full Names']
```

```python
afullnames = df['Author Full Names'].tolist()
```

```python
[name.strip() for name in afullnames[0].split(';')]
```

```python
authors_abv= []

for fullnames in afullnames:
    authors_abv.extend([name.strip() for name in fullnames.split(';')])
```

```python
def original_name(abbreviation):
    return ' '.join([name.strip() for name in reversed(abbreviation.split(','))])

original_name(authors_abv[0])
```

```python
authors = [original_name(author) for author in authors_abv]
```

```python
from collections import Counter
```

```python
authors_count = Counter(authors)
authors_count
```

```python
authornames = authors_count.keys()
counts = [authors_count[author] for author in authornames]

authorwise_df = pd.DataFrame({"Author" : authornames, "Publications" : counts})
```

```python
authorwise_df
```

```python
authorwise_df = authorwise_df.sort_values(by='Publications',ascending=False)
authorwise_df.head(10)
```
