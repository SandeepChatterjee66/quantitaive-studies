```python
import os
import pandas as pd
```

```python
path = os.getcwd()+'/dataset'
files = os.listdir(path)
files
```

```python
os.chdir(path)
files_xls = [f for f in files if f[-3:] == 'xls']
files_xls
```

```python
dfs = []
```

```python
for f in files_xls:
    data = pd.read_excel(f)
    dfs.append(data)
```

```python
df = pd.concat(dfs)
```
